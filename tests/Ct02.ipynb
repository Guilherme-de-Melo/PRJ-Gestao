```python
# Validação de campos obrigatória
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

servico = Service(ChromeDriverManager().install())

navegador = webdriver.Chrome(service=servico) 
```

```python
# acessa página
navegador.get("http://127.0.0.1:5500/Front%20end/index.html")
```

```python
# Clicar direto para fazer a requisição sem selecionar campos
navegador.find_element('xpath', '/html/body/main/div/div/form/div[5]/button').click()
```
