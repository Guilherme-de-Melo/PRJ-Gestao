```python
# fluxo completo
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service

servico = Service(ChromeDriverManager().install())

navegador = webdriver.Chrome(service=servico) 
```

```python
# acessa página
navegador.get("http://127.0.0.1:5500/Front%20end/index.html")
```

```python
# seleciona o fabricante do processador
navegador.find_element('xpath', '/html/body/main/div/div/form/div[1]/button[1]').click()
```

```python
# abre o listBox dos modelos de processador
navegador.find_element('xpath', '/html/body/main/div/div/form/div[2]/div/div').click()
```

```python
# seleciona o modelo de procesador
navegador.find_element('xpath', '/html/body/main/div/div/form/div[2]/ul/li[2]').click()
```

```python
# seleciona o fabricante do Placa de Video
navegador.find_element('xpath', '/html/body/main/div/div/form/div[3]/button[1]').click()
```

```python
# abre o listBox dos modelos de Placa de Video
navegador.find_element('xpath', '/html/body/main/div/div/form/div[4]/div/div').click()
```

```python
# seleciona o modelo de Placa de Video
navegador.find_element('xpath', '/html/body/main/div/div/form/div[4]/ul/li[3]').click()
```

```python
# faz requisição na api
navegador.find_element('xpath', '/html/body/main/div/div/form/div[5]/button').click()
```
